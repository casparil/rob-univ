```python
results_base_dir = "/root/univ-data/results"

results = {
    "imagenet100": {
        "standard":{
            "dis": {
                "eps0":f"{results_base_dir}/imagenet100/standard/dis_eps0_0.csv",
                "eps025":f"{results_base_dir}/imagenet100/standard/dis_eps025_0.csv",
                "eps05":f"{results_base_dir}/imagenet100/standard/dis_eps05_0.csv",
                "eps1":f"{results_base_dir}/imagenet100/standard/dis_eps1_0.csv",
                "eps3":f"{results_base_dir}/imagenet100/standard/dis_eps3_0.csv",
            },
            "jsd": {
                "eps0":f"{results_base_dir}/imagenet100/standard/jsd_eps0_0.csv",
                "eps025":f"{results_base_dir}/imagenet100/standard/jsd_eps025_0.csv",
                "eps05":f"{results_base_dir}/imagenet100/standard/jsd_eps05_0.csv",
                "eps1":f"{results_base_dir}/imagenet100/standard/jsd_eps1_0.csv",
                "eps3":f"{results_base_dir}/imagenet100/standard/jsd_eps3_0.csv",
            },
            "cka": {
                "eps0":f"{results_base_dir}/imagenet100/standard/cka_eps0_0.1.csv",
                "eps025":f"{results_base_dir}/imagenet100/standard/cka_eps025_0.1.csv",
                "eps05":f"{results_base_dir}/imagenet100/standard/cka_eps05_0.1.csv",
                "eps1":f"{results_base_dir}/imagenet100/standard/cka_eps1_0.1.csv",
                "eps3":f"{results_base_dir}/imagenet100/standard/cka_eps3_0.1.csv",
            },
            "jaccard": {
                "eps0":f"{results_base_dir}/imagenet100/standard/jac_mean_eps0_0.1.csv",
                "eps025":f"{results_base_dir}/imagenet100/standard/jac_mean_eps025_0.1.csv",
                "eps05":f"{results_base_dir}/imagenet100/standard/jac_mean_eps05_0.1.csv",
                "eps1":f"{results_base_dir}/imagenet100/standard/jac_mean_eps1_0.1.csv",
                "eps3":f"{results_base_dir}/imagenet100/standard/jac_mean_eps3_0.1.csv",
            },
            "proc": {
                "eps0":f"{results_base_dir}/imagenet100/standard/proc_norm_mean_eps0_0.1.csv",
                "eps025":f"{results_base_dir}/imagenet100/standard/proc_norm_mean_eps025_0.1.csv",
                "eps05":f"{results_base_dir}/imagenet100/standard/proc_norm_mean_eps05_0.1.csv",
                "eps1":f"{results_base_dir}/imagenet100/standard/proc_norm_mean_eps1_0.1.csv",
                "eps3":f"{results_base_dir}/imagenet100/standard/proc_norm_mean_eps3_0.1.csv",
            },
            "gulp": {
                "eps0":f"{results_base_dir}/imagenet100/standard/gulp_mean_eps0_0.1.csv",
                "eps025":f"{results_base_dir}/imagenet100/standard/gulp_mean_eps025_0.1.csv",
                "eps05":f"{results_base_dir}/imagenet100/standard/gulp_mean_eps05_0.1.csv",
                "eps1":f"{results_base_dir}/imagenet100/standard/gulp_mean_eps1_0.1.csv",
                "eps3":f"{results_base_dir}/imagenet100/standard/gulp_mean_eps3_0.1.csv",
            },
            "rtd": {
                "eps0":f"{results_base_dir}/imagenet100/standard/rtd_mean_eps0_0.1.csv",
                "eps025":f"{results_base_dir}/imagenet100/standard/rtd_mean_eps025_0.1.csv",
                "eps05":f"{results_base_dir}/imagenet100/standard/rtd_mean_eps05_0.1.csv",
                "eps1":f"{results_base_dir}/imagenet100/standard/rtd_mean_eps1_0.1.csv",
                "eps3":f"{results_base_dir}/imagenet100/standard/rtd_mean_eps3_0.1.csv",
            },
        },
        "inverted": {
            "dis": {
                "eps0":f"{results_base_dir}/imagenet100/dis_eps0_0.csv",
                "eps025":f"{results_base_dir}/imagenet100/dis_eps025_0.csv",
                "eps05":f"{results_base_dir}/imagenet100/dis_eps05_0.csv",
                "eps1":f"{results_base_dir}/imagenet100/dis_eps1_0.csv",
                "eps3":f"{results_base_dir}/imagenet100/dis_eps3_0.csv",
            },
            "jsd": {
                "eps0":f"{results_base_dir}/imagenet100/jsd_eps0_0.csv",
                "eps025":f"{results_base_dir}/imagenet100/jsd_eps025_0.csv",
                "eps05":f"{results_base_dir}/imagenet100/jsd_eps05_0.csv",
                "eps1":f"{results_base_dir}/imagenet100/jsd_eps1_0.csv",
                "eps3":f"{results_base_dir}/imagenet100/jsd_eps3_0.csv",
            },
            "cka": {
                "eps0":f"{results_base_dir}/imagenet100/cka_eps0_0.1.csv",
                "eps025":f"{results_base_dir}/imagenet100/cka_eps025_0.1.csv",
                "eps05":f"{results_base_dir}/imagenet100/cka_eps05_0.1.csv",
                "eps1":f"{results_base_dir}/imagenet100/cka_eps1_0.1.csv",
                "eps3":f"{results_base_dir}/imagenet100/cka_eps3_0.1.csv",
            },
            "jaccard": {
                "eps0":f"{results_base_dir}/imagenet100/jac_mean_eps0_0.1.csv",
                "eps025":f"{results_base_dir}/imagenet100/jac_mean_eps025_0.1.csv",
                "eps05":f"{results_base_dir}/imagenet100/jac_mean_eps05_0.1.csv",
                "eps1":f"{results_base_dir}/imagenet100/jac_mean_eps1_0.1.csv",
                "eps3":f"{results_base_dir}/imagenet100/jac_mean_eps3_0.1.csv",
            },
            "proc": {
                "eps0":f"{results_base_dir}/imagenet100/proc_norm_mean_eps0_0.1.csv",
                "eps025":f"{results_base_dir}/imagenet100/proc_norm_mean_eps025_0.1.csv",
                "eps05":f"{results_base_dir}/imagenet100/proc_norm_mean_eps05_0.1.csv",
                "eps1":f"{results_base_dir}/imagenet100/proc_norm_mean_eps1_0.1.csv",
                "eps3":f"{results_base_dir}/imagenet100/proc_norm_mean_eps3_0.1.csv",
            },
            "gulp": {
                "eps0":f"{results_base_dir}/imagenet100/gulp_mean_eps0_0.1.csv",
                "eps025":f"{results_base_dir}/imagenet100/gulp_mean_eps025_0.1.csv",
                "eps05":f"{results_base_dir}/imagenet100/gulp_mean_eps05_0.1.csv",
                "eps1":f"{results_base_dir}/imagenet100/gulp_mean_eps1_0.1.csv",
                "eps3":f"{results_base_dir}/imagenet100/gulp_mean_eps3_0.1.csv",
            },
            "rtd": {
                "eps0":f"{results_base_dir}/imagenet100/rtd_mean_eps0_0.1.csv",
                "eps025":f"{results_base_dir}/imagenet100/rtd_mean_eps025_0.1.csv",
                "eps05":f"{results_base_dir}/imagenet100/rtd_mean_eps05_0.1.csv",
                "eps1":f"{results_base_dir}/imagenet100/rtd_mean_eps1_0.1.csv",
                "eps3":f"{results_base_dir}/imagenet100/rtd_mean_eps3_0.1.csv",
            },
        }
    },
    "imagenet1k": {
        "inverted": {
            "dis": {
                "eps0":f"{results_base_dir}/imagenet1k/dis_eps0_1.csv",
                "eps025":f"{results_base_dir}/imagenet1k/dis_eps025_1.csv",
                "eps05":f"{results_base_dir}/imagenet1k/dis_eps05_1.csv",
                "eps1":f"{results_base_dir}/imagenet1k/dis_eps1_1.csv",
                "eps3":f"{results_base_dir}/imagenet1k/dis_eps3_1.csv",
            },
            "jsd": {
                "eps0":f"{results_base_dir}/imagenet1k/jsd_eps0_1.csv",
                "eps025":f"{results_base_dir}/imagenet1k/jsd_eps025_1.csv",
                "eps05":f"{results_base_dir}/imagenet1k/jsd_eps05_1.csv",
                "eps1":f"{results_base_dir}/imagenet1k/jsd_eps1_1.csv",
                "eps3":f"{results_base_dir}/imagenet1k/jsd_eps3_1.csv",
            },
            "cka": {
                "eps0":f"{results_base_dir}/imagenet1k/cka_eps0_0.1.csv",
                "eps025":f"{results_base_dir}/imagenet1k/cka_eps025_0.1.csv",
                "eps05":f"{results_base_dir}/imagenet1k/cka_eps05_0.1.csv",
                "eps1":f"{results_base_dir}/imagenet1k/cka_eps1_0.1.csv",
                "eps3":f"{results_base_dir}/imagenet1k/cka_eps3_0.1.csv",
            },
            "jaccard": {
                "eps0":f"{results_base_dir}/imagenet1k/jac_mean_eps0_0.1.csv",
                "eps025":f"{results_base_dir}/imagenet1k/jac_mean_eps025_0.1.csv",
                "eps05":f"{results_base_dir}/imagenet1k/jac_mean_eps05_0.1.csv",
                "eps1":f"{results_base_dir}/imagenet1k/jac_mean_eps1_0.1.csv",
                "eps3":f"{results_base_dir}/imagenet1k/jac_mean_eps3_0.1.csv",
            },
            "proc": {
                "eps0":f"{results_base_dir}/imagenet1k/proc_norm_mean_eps0_0.1.csv",
                "eps025":f"{results_base_dir}/imagenet1k/proc_norm_mean_eps025_0.1.csv",
                "eps05":f"{results_base_dir}/imagenet1k/proc_norm_mean_eps05_0.1.csv",
                "eps1":f"{results_base_dir}/imagenet1k/proc_norm_mean_eps1_0.1.csv",
                "eps3":f"{results_base_dir}/imagenet1k/proc_norm_mean_eps3_0.1.csv",
            },
            "gulp": {
                "eps0":f"{results_base_dir}/imagenet1k/gulp_mean_eps0_0.1.csv",
                "eps025":f"{results_base_dir}/imagenet1k/gulp_mean_eps025_0.1.csv",
                "eps05":f"{results_base_dir}/imagenet1k/gulp_mean_eps05_0.1.csv",
                "eps1":f"{results_base_dir}/imagenet1k/gulp_mean_eps1_0.1.csv",
                "eps3":f"{results_base_dir}/imagenet1k/gulp_mean_eps3_0.1.csv",
            },
            "rtd": {
                "eps0":f"{results_base_dir}/imagenet1k/rtd_mean_eps0_0.1.csv",
                "eps025":f"{results_base_dir}/imagenet1k/rtd_mean_eps025_0.1.csv",
                "eps05":f"{results_base_dir}/imagenet1k/rtd_mean_eps05_0.1.csv",
                "eps1":f"{results_base_dir}/imagenet1k/rtd_mean_eps1_0.1.csv",
                "eps3":f"{results_base_dir}/imagenet1k/rtd_mean_eps3_0.1.csv",
            },
        },
        "standard": {
            "dis": {
                "eps0":f"{results_base_dir}/imagenet1k/standard/dis_eps0_0.csv",
                "eps025":f"{results_base_dir}/imagenet1k/standard/dis_eps025_0.csv",
                "eps05":f"{results_base_dir}/imagenet1k/standard/dis_eps05_0.csv",
                "eps1":f"{results_base_dir}/imagenet1k/standard/dis_eps1_0.csv",
                "eps3":f"{results_base_dir}/imagenet1k/standard/dis_eps3_0.csv",
            },
            "jsd": {
                "eps0":f"{results_base_dir}/imagenet1k/standard/jsd_eps0_1.csv",
                "eps025":f"{results_base_dir}/imagenet1k/standard/jsd_eps025_1.csv",
                "eps05":f"{results_base_dir}/imagenet1k/standard/jsd_eps05_1.csv",
                "eps1":f"{results_base_dir}/imagenet1k/standard/jsd_eps1_1.csv",
                "eps3":f"{results_base_dir}/imagenet1k/standard/jsd_eps3_1.csv",
            },
            "cka": {
                "eps0":f"{results_base_dir}/imagenet1k/standard/cka_eps0_0.1.csv",
                "eps025":f"{results_base_dir}/imagenet1k/standard/cka_eps025_0.1.csv",
                "eps05":f"{results_base_dir}/imagenet1k/standard/cka_eps05_0.1.csv",
                "eps1":f"{results_base_dir}/imagenet1k/standard/cka_eps1_0.1.csv",
                "eps3":f"{results_base_dir}/imagenet1k/standard/cka_eps3_0.1.csv",
            },
            "jaccard": {
                "eps0":f"{results_base_dir}/imagenet1k/standard/jac_mean_eps0_0.1.csv",
                "eps025":f"{results_base_dir}/imagenet1k/standard/jac_mean_eps025_0.1.csv",
                "eps05":f"{results_base_dir}/imagenet1k/standard/jac_mean_eps05_0.1.csv",
                "eps1":f"{results_base_dir}/imagenet1k/standard/jac_mean_eps1_0.1.csv",
                "eps3":f"{results_base_dir}/imagenet1k/standard/jac_mean_eps3_0.1.csv",
            },
            "proc": {
                "eps0":f"{results_base_dir}/imagenet1k/standard/proc_norm_mean_eps0_0.1.csv",
                "eps025":f"{results_base_dir}/imagenet1k/standard/proc_norm_mean_eps025_0.1.csv",
                "eps05":f"{results_base_dir}/imagenet1k/standard/proc_norm_mean_eps05_0.1.csv",
                "eps1":f"{results_base_dir}/imagenet1k/standard/proc_norm_mean_eps1_0.1.csv",
                "eps3":f"{results_base_dir}/imagenet1k/standard/proc_norm_mean_eps3_0.1.csv",
            },
            "gulp": {
                "eps0":f"{results_base_dir}/imagenet1k/standard/gulp_mean_eps0_0.1.csv",
                "eps025":f"{results_base_dir}/imagenet1k/standard/gulp_mean_eps025_0.1.csv",
                "eps05":f"{results_base_dir}/imagenet1k/standard/gulp_mean_eps05_0.1.csv",
                "eps1":f"{results_base_dir}/imagenet1k/standard/gulp_mean_eps1_0.1.csv",
                "eps3":f"{results_base_dir}/imagenet1k/standard/gulp_mean_eps3_0.1.csv",
            },
            "rtd": {
                "eps0":f"{results_base_dir}/imagenet1k/standard/rtd_mean_eps0_0.1.csv",
                "eps025":f"{results_base_dir}/imagenet1k/standard/rtd_mean_eps025_0.1.csv",
                "eps05":f"{results_base_dir}/imagenet1k/standard/rtd_mean_eps05_0.1.csv",
                "eps1":f"{results_base_dir}/imagenet1k/standard/rtd_mean_eps1_0.1.csv",
                "eps3":f"{results_base_dir}/imagenet1k/standard/rtd_mean_eps3_0.1.csv",
            },
        }
    },
    "cifar10": {
        "inverted": {
            "dis": {
                "eps0": f"{results_base_dir}/cifar10/dis_eps0_0.csv",
                "eps025": f"{results_base_dir}/cifar10/dis_eps025_0.csv",
                "eps05": f"{results_base_dir}/cifar10/dis_eps05_0.csv",
                "eps1": f"{results_base_dir}/cifar10/dis_eps1_0.csv",
            },
            "jsd": {
                "eps0":f"{results_base_dir}/cifar10/jsd_eps0_0.csv",
                "eps025":f"{results_base_dir}/cifar10/jsd_eps025_0.csv",
                "eps05":f"{results_base_dir}/cifar10/jsd_eps05_0.csv",
                "eps1":f"{results_base_dir}/cifar10/jsd_eps1_0.csv",
            },
            "cka": {
                "eps0":f"{results_base_dir}/cifar10/cka_eps0_0.1.csv",
                "eps025":f"{results_base_dir}/cifar10/cka_eps025_0.1.csv",
                "eps05":f"{results_base_dir}/cifar10/cka_eps05_0.1.csv",
                "eps1":f"{results_base_dir}/cifar10/cka_eps1_0.1.csv",
            },
            "2ndcos": {
                "eps0":f"{results_base_dir}/cifar10/cos_sim_mean_eps0_0.csv",
                "eps025":f"{results_base_dir}/cifar10/cos_sim_mean_eps025_0.csv",
                "eps05":f"{results_base_dir}/cifar10/cos_sim_mean_eps05_0.csv",
                "eps1":f"{results_base_dir}/cifar10/cos_sim_mean_eps1_0.csv",
            },
            "jaccard": {
                "eps0":f"{results_base_dir}/cifar10/jac_eps0_0.1.csv",
                "eps025":f"{results_base_dir}/cifar10/jac_eps025_0.1.csv",
                "eps05":f"{results_base_dir}/cifar10/jac_eps05_0.1.csv",
                "eps1":f"{results_base_dir}/cifar10/jac_eps1_0.1.csv",
            },
            "proc": {
                "eps0":f"{results_base_dir}/cifar10/proc_norm_mean_eps0_0.1.csv",
                "eps025":f"{results_base_dir}/cifar10/proc_norm_mean_eps025_0.1.csv",
                "eps05":f"{results_base_dir}/cifar10/proc_norm_mean_eps05_0.1.csv",
                "eps1":f"{results_base_dir}/cifar10/proc_norm_mean_eps1_0.1.csv",
            },
            "rtd": {
                "eps0":f"{results_base_dir}/cifar10/rtd_mean_eps0_0.1.csv",
                "eps025":f"{results_base_dir}/cifar10/rtd_mean_eps025_0.1.csv",
                "eps05":f"{results_base_dir}/cifar10/rtd_mean_eps05_0.1.csv",
                "eps1":f"{results_base_dir}/cifar10/rtd_mean_eps1_0.1.csv",
            },
        },
        "standard": {
            "dis": {
                "eps0": f"{results_base_dir}/cifar10/standard/dis_eps0_0.csv",
                "eps025": f"{results_base_dir}/cifar10/standard/dis_eps025_0.csv",
                "eps05": f"{results_base_dir}/cifar10/standard/dis_eps05_0.csv",
                "eps1": f"{results_base_dir}/cifar10/standard/dis_eps1_0.csv",
            }
            ,
            "jsd": {
                "eps0":f"{results_base_dir}/cifar10/standard/jsd_eps0_0.csv",
                "eps025":f"{results_base_dir}/cifar10/standard/jsd_eps025_0.csv",
                "eps05":f"{results_base_dir}/cifar10/standard/jsd_eps05_0.csv",
                "eps1":f"{results_base_dir}/cifar10/standard/jsd_eps1_0.csv",
            },
            "cka": {
                "eps0":f"{results_base_dir}/cifar10/standard/cka_eps0_0.1.csv",
                "eps025":f"{results_base_dir}/cifar10/standard/cka_eps025_0.1.csv",
                "eps05":f"{results_base_dir}/cifar10/standard/cka_eps05_0.1.csv",
                "eps1":f"{results_base_dir}/cifar10/standard/cka_eps1_0.1.csv",
            },
            "jaccard": {
                "eps0":f"{results_base_dir}/cifar10/standard/jac_eps0_0.1.csv",
                "eps025":f"{results_base_dir}/cifar10/standard/jac_eps025_0.1.csv",
                "eps05":f"{results_base_dir}/cifar10/standard/jac_eps05_0.1.csv",
                "eps1":f"{results_base_dir}/cifar10/standard/jac_eps1_0.1.csv",
            },
            "proc": {
                "eps0":f"{results_base_dir}/cifar10/standard/proc_norm_mean_eps0_0.1.csv",
                "eps025":f"{results_base_dir}/cifar10/standard/proc_norm_mean_eps025_0.1.csv",
                "eps05":f"{results_base_dir}/cifar10/standard/proc_norm_mean_eps05_0.1.csv",
                "eps1":f"{results_base_dir}/cifar10/standard/proc_norm_mean_eps1_0.1.csv",
            },
            "rtd": {
                "eps0":f"{results_base_dir}/cifar10/standard/rtd_mean_eps0_0.1.csv",
                "eps025":f"{results_base_dir}/cifar10/standard/rtd_mean_eps025_0.1.csv",
                "eps05":f"{results_base_dir}/cifar10/standard/rtd_mean_eps05_0.1.csv",
                "eps1":f"{results_base_dir}/cifar10/standard/rtd_mean_eps1_0.1.csv",
            },
        }
    },
}


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats

sns.set_theme("paper", style="darkgrid", font_scale=1.5)


def eps_to_float(eps: str) -> float:
    if eps == "eps0":
        return 0.0
    elif eps == "eps025":
        return 0.25
    elif eps == "eps05":
        return 0.5
    elif eps == "eps1":
        return 1.0
    elif eps == "eps3":
        return 3.0
    else:
        raise ValueError(f"Eps level not recognized: {eps}")

dfs = []
for dataset, dataset_results in results.items():
    for input_type, input_type_results in dataset_results.items():
        for measure, path_dict in input_type_results.items():
            for eps, csv_path in path_dict.items():
                try:
                    df = pd.read_csv(csv_path, index_col=0)
                except FileNotFoundError as e:
                    print(e)
                    continue

                # Mask out comparisons between identical models, so we can drop them easier later
                for index_val in df.index:
                    # df.loc[index_val, index_val] = 1.0
                    df.loc[index_val, index_val] = np.NaN

                # Give name to index
                df = df.rename_axis("model1")

                # Bring data to long format (many rows) and remove self comparisons
                df = df.reset_index().melt(id_vars="model1", var_name="model2", value_name="score")
                df = df.dropna(axis=0)

                df["measure"] = measure
                df["eps"] = eps_to_float(eps)
                df["dataset"] = dataset
                df["input"] = input_type

                if measure == "proc":
                    df.loc[:, "score"] = (2 - df.loc[:, "score"]) / 2

                if measure == "dis":
                    df.loc[:, "score"] = 1 - df.loc[:, "score"]

                if measure == "jsd":
                    df.loc[:, "score"] = (np.log(2) - df.loc[:, "score"]) / np.log(2)

                if measure == "rtd":
                    df.loc[:, "score"] = -1 * df.loc[:, "score"]

                dfs.append(df)

data = pd.concat(dfs, axis=0).reset_index(drop=True)

# Exclude tiny_vit_5m results on ImageNet100 and 50, because we mistakingly used checkpoints for IN1k
data = data.loc[~(((data.model1 == "tiny_vit_5m") | (data.model2 == "tiny_vit5m")) & (data.dataset.isin(["imagenet100", "imagenet50"])))]

# Exclude broken densenet161 on IN1k
# data = data.loc[~(((data.model1 == "densenet161") | (data.model2 == "densenet161")) & (data.dataset.isin(["imagenet1k"])))]


data.head()
```

### Agreement
#### Standard

```python
def disagreement_lower_bound(x, num_classes: int=1000):
    """Fort et al 2020. Deep Ensembles: A Loss Landscape Perspective. Apx D2"""
    a = x["acc1"]/100
    a_star = x["acc2"]/100

    if np.isnan(a) or np.isnan(a_star):
        return np.nan

    if a > a_star:
        a_star, a = a, a_star

    return (num_classes - 1) * (a_star - a) / (num_classes * a_star - 1)


def disagreement_upper_bound(x, num_classes: int=1000):
    """Fort et al 2020. Deep Ensembles: A Loss Landscape Perspective. Apx D1"""
    a = x["acc1"]/100
    a_star = x["acc2"]/100

    if np.isnan(a) or np.isnan(a_star):
        return np.nan

    if a > a_star:
        a_star, a = a, a_star

    return (1 - a_star) * a + (1 - a) * a_star + (1 - a_star) * (1 - a) * (num_classes - 2) / (num_classes - 1)


def combine_agreement_with_acc(data: pd.DataFrame, evals: pd.DataFrame, dataset: str, num_classes: int = 1000, input_type="standard", combine_with_regular_acc: bool=True):
    agreement_data = data.loc[(data.dataset == dataset) & (data.measure == "dis") & (data.input == input_type), :].copy()
    if combine_with_regular_acc:
        evals_subset = evals.loc[(evals.dataset == dataset) & (evals.attack_eps==0), :]
    else:
        evals_subset = evals.loc[(evals.dataset == dataset) & (evals.attack_eps==evals.trained_eps), :]


    evals_subset = evals_subset.drop_duplicates(["model", "dataset", "acc", "trained_eps", "attack_lr", "attack_eps"], keep="last")
    print(len(agreement_data), len(evals_subset))

    # Add accuracy information to each comparison
    merged = pd.merge(
        agreement_data,
        evals_subset.rename(columns={"model": "model1", "trained_eps": "eps"}),
        "left",
        on=["model1", "eps"],
        suffixes=(None, "_y"),
    )
    merged = merged.drop(["loss", "attack_lr", "attack_eps", "timestamp", "dataset_y"], axis="columns")
    merged = merged.rename(columns={"acc": "acc1"})

    merged = pd.merge(
        merged,
        evals_subset.rename(columns={"model": "model2", "trained_eps": "eps"}),
        "left",
        on=["model2", "eps"],
        suffixes=(None, "_y"),
    )
    merged = merged.drop(["loss", "attack_lr", "attack_eps", "timestamp", "dataset_y"], axis="columns")
    merged = merged.rename(columns={"acc": "acc2"})

    # Add bounds
    merged["ub"] = 1 - merged.apply(disagreement_lower_bound, axis=1, num_classes=num_classes)
    merged["lb"] = 1 - merged.apply(disagreement_upper_bound, axis=1, num_classes=num_classes)
    return merged



n_panels = 2
fig, axes = plt.subplots(1, n_panels, figsize=(n_panels*3*1.61, 3))

fontsize=24
# fontweight="normal"
fontweight="bold"

ax_idx = 0
ax = axes[ax_idx]
plotdata = combine_agreement_with_acc(data2, evals, "imagenet1k")
sns.boxplot(data=plotdata, x="eps", y="score", ax=ax, **color_kwargs)
ax.set_title("ImageNet1k")
ax.set_ylabel("Agreement")
ax.set_xlabel(r"Robustness $\epsilon$")
bounds = plotdata.groupby(["eps"])[["ub", "lb"]].mean()
ax.plot([0, 1, 2, 3, 4], bounds["lb"], color="gray", linestyle="--")
ax.plot([0, 1, 2, 3, 4], bounds["ub"], color="gray", linestyle="-.")
ax.text(3.4, 0.94, "Avg. Upper Bound", color="gray", ha="center")
ax.text(3.4, 0.51, "Avg. Lower Bound", color="gray", ha="center")
ax.text(x=-0.15, y=1.03, s="A", va="bottom", color="black", transform=ax.transAxes, fontdict={
                        'fontsize': fontsize,
                        'fontweight': fontweight
                    })
ax_idx += 1

# ax = axes[ax_idx]
# plotdata = combine_agreement_with_acc(data2, evals, "imagenet100", num_classes=100)
# sns.boxplot(data=plotdata, x="eps", y="score", ax=ax, **color_kwargs)
# ax.set_title("ImageNet100")
# ax.set_ylabel("Agreement")
# ax.set_xlabel(r"Robustness $\epsilon$")
# bounds = plotdata.groupby(["eps"])[["ub", "lb"]].mean()
# ax.plot([0, 1, 2, 3, 4], bounds["lb"], color="gray", linestyle="--")
# ax.plot([0, 1, 2, 3, 4], bounds["ub"], color="gray", linestyle="-.")
# ax_idx += 1

ax = axes[ax_idx]
plotdata = combine_agreement_with_acc(data2, evals, "cifar10", num_classes=10)
sns.boxplot(data=plotdata, x="eps", y="score", ax=ax, **color_kwargs)
ax.set_title("CIFAR10")
ax.set_ylabel("Agreement")
ax.set_xlabel(r"Robustness $\epsilon$")
bounds = plotdata.groupby(["eps"])[["ub", "lb"]].mean()
ax.plot([0, 1, 2, 3], bounds["lb"], color="gray", linestyle="--")
ax.plot([0, 1, 2, 3], bounds["ub"], color="gray", linestyle="-.")
ax.text(2.8, 0.89, "Avg. Upper Bound", color="gray", ha="center")
ax.text(2.5, 0.47, "Avg. Lower Bound", color="gray", ha="right")


fig.savefig("../figs/agreement_regular.pdf", bbox_inches="tight")
```
